```python
import pandas as pd
import re
import numpy as np
import sys

df = pd.read_csv('tweets.csv')

tweets = df['text']

replypattern = '@[\w]+'
urlpattern = 'https?://[\w/:%#\$&\?\(\)~\.=\+\-]+'
enterpattern = "\n"
spacepattern = " "
hashpattern = "#"

#processedtweets = []
counting = 0
for tweet in tweets:
    counting += 1
    if counting == 20000:
        break
    i = re.sub(replypattern, '', tweet)
    i = re.sub(enterpattern, '', i)
    i = re.sub(urlpattern, '' , i)
    i = re.sub(spacepattern, '' , i)
    i = re.sub(hashpattern, '' , i)
    if isinstance(i, str) and not i.split():
        pass
    else:
        sys.stdout.write('\r前処理中 {}'.format(counting))
        f = open('./tweets/tweets/onigirichanre'+str(counting)+'.txt','w')
        f.write(i)
        f.close()
        #processedtweets.append(i)
        

#processedtweetsDataFrame = pd.Series(processedtweets)
#newDF = pd.DataFrame({'text': processedtweetsDataFrame})

#newDF.to_csv('processedtweets.csv',index=False)
```
