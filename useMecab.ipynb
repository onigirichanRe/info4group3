```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
import pandas as pd
import re

df = pd.read_csv('tweets.csv')

tweets = df['text']

#URL等のノイズを除去
replypattern = '@[\w]+'
urlpattern = 'https?://[\w/:%#\$&\?\(\)~\.=\+\-]+'

processedtweets = []
counting = 1
for tweet in tweets:
    counting += 1
    if counting == 100: #読み込むツイートはとりあえず100個だけ
        break
    i = re.sub(replypattern, '', tweet)
    i = re.sub(urlpattern, '', i)
    if isinstance(i, str) and not i.split():
        pass
    else:
        processedtweets.append(i)

processedtweetsDataFrame = pd.Series(processedtweets)
newDF = pd.DataFrame({'text': processedtweetsDataFrame})

newDF.to_csv('processedtweets.csv',index=False)
```

```python
#リストbygramに先ほどの整形済みツイートデータを入れる
f = open('processedtweets.csv')

line = f.readline()
bygram = []
while line:
    line = f.readline()
    bygram.append(line)
    #for e in ngram(text, 2):
     #   bygram.append(e)
f.close()
```

```python
#上でbygramに入れたツイートをMeCabにかける
import nltk
import MeCab
jp_sent_tokenizer = nltk.RegexpTokenizer('[^　「」！？。]*[！？。]')
f = open('ketai.txt','w')

docs = []

for i in range(len(bygram)):
    sents = jp_sent_tokenizer.tokenize( bygram[i] )
    #print('# doc -> sentences分割処理の例。')
    mecab = MeCab.Tagger()
    mecab.parse(bygram[i])
    node = mecab.parseToNode(bygram[i])
    while node:
        docs.append(node.surface)
        node = node.next

for x in docs:
    print(x)
    f.write(str(x) + "\n")

f.close()
```
